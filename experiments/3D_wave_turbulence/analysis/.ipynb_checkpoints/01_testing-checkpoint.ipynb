```python
import xarray as xr
import numpy as np
import xgcm
from xmitgcm import open_mdsdataset
from matplotlib import pyplot as plt
import matplotlib.colors as colors
%matplotlib inline
```

```python
data_dir = "../run"
ds = open_mdsdataset(data_dir,iters=[120,43200],ignore_unknown_vars=True,prefix=["standardDiag"])
ds['time'] = ds['time']*30 # convert from iterations to seconds
ds['time'].attrs['units'] = "seconds since 2000-1-1 0:0:0"
grid = xgcm.Grid(ds, periodic=['X','Y'])

coords = ds.coords.to_dataset().reset_coords()
ds = ds.reset_coords(drop=True)
coords['dV'] = coords['rA']*(coords['drF']*coords['hFacC'])
```

```python
nu = 2.e-3
# Calculate contributions to dissipation on tracer grid
ds['EPSILON'] = (nu
                 # (du/dx)**2=((du*dy*dz)/dV)**2
                 *(
                 (grid.diff(ds['UVEL']
                          *coords['dyG']*(coords['drF']*coords['hFacW']),axis='X')
                  /coords['dV'])**2
                 # (dv/dx)**2=((dv*dy*dz)/dV)**2
                 +(grid.diff(grid.interp(grid.interp(ds['VVEL'],axis='X'),axis='Y')
                         *coords['dyG']*(coords['drF']*coords['hFacW']),axis='X')
                  /coords['dV'])**2
                 # (dw/dx)**2=((dw*dy*dz)/dV)**2
                 +(grid.diff(grid.interp(grid.interp(ds['WVEL'],axis='X'),axis='Z',boundary='fill')
                          *coords['dyG']*(coords['drF']*coords['hFacW']),axis='X')
                  /coords['dV'])**2
                 # (du/dy)**2=((du*dx*dz)/dV)**2
                 +(grid.diff(grid.interp(grid.interp(ds['UVEL'],axis='X'),axis='Y')
                           *coords['dxG']*(coords['drF']*coords['hFacS']),axis='Y')
                  /coords['dV'])**2
                 # (dv/dy)**2=((dv*dx*dz)/dV)**2
                 +(grid.diff(ds['VVEL']
                          *coords['dxG']*(coords['drF']*coords['hFacS']),axis='Y')
                  /coords['dV'])**2
                 # (dw/dy)**2=((dw*dx*dz)/dV)**2
                 +(grid.diff(grid.interp(grid.interp(ds['WVEL'],axis='Y'),axis='Z',boundary='fill')
                           *coords['dxG']*(coords['drF']*coords['hFacS']),axis='Y')
                  /coords['dV'])**2
                 # (du/dz)**2=((du*dA)/dV)**2
                 +(grid.diff(grid.interp(grid.interp(ds['UVEL'],axis='Z',boundary='fill'),axis='X')
                          *coords['rA'],axis='Z',boundary='fill')
                  /coords['dV'])**2
                 # (dv/dz)**2=((dv*dA)/dV)**2
                 +(grid.diff(grid.interp(grid.interp(ds['VVEL'],axis='Z',boundary='fill'),axis='Y')
                           *coords['rA'],axis='Z',boundary='fill')
                  /coords['dV'])**2
                 # (dw/dz)**2=((dw*dA)/dV)**2
                 +(grid.diff(ds['WVEL']
                            *coords['rA'],axis='Z',boundary='fill')
                  /coords['dV'])**2
                ))
ds['EPSILON'].load()
```

```python
q = ds['EPSILON'].sel(time=ds['time'][-1],XC=ds['XC'][0]).plot(cmap=plt.get_cmap("viridis"),norm=colors.LogNorm(vmin=1.e-11,vmax=1.e-8))
q.set_clim([1.e-13,1.e-11])
```

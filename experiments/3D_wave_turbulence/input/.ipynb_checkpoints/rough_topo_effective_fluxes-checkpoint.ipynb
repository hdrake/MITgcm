```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from mpl_toolkits.axes_grid.inset_locator import inset_axes
%matplotlib inline

# topography
h0 = 110 # rms height
k0 = 2.2e-4 # roll-off wavenumber
l0 = 1e-3 # roll-off wavenumber
nu = 0.9 # high wavenumber power law
N = 1e-3

n = 500
Tk = 1e2
Tl = 1e1
k = np.linspace(0,1.0/(2.0*Tk),n//2)
l = np.linspace(0,1.0/(2.0*Tl),n//2)
k,l = np.meshgrid(k,l)

H = (4*np.pi*nu*h0**2)/(l0*k0*((k/k0)**2 + (l/l0)**2 + 1.0)**(nu+1)) # topographic spectrum
H = H*np.exp(1j*np.random.random((n//2,n//2))*2*np.pi)
H[n//2+1:,:]=0
H[:,n//2+1:]=0

h = (2.0/n)**2*np.fft.ifft2(H,(n//2,n//2))
x_ = np.linspace(0,n*Tk,n//2)
y_ = np.linspace(0,n*Tl,n//2)
x,y = np.meshgrid(x_,y_)

```

```python
plt.pcolor(k,l,np.abs(H))
plt.colorbar()
plt.ylim([0,0.05])
plt.xlim([0,0.005])
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(x[0:200,0:200]/1000.,y[0:200,0:200]/1000.,h.real[0:200,0:200],linewidth=0,cmap=cm.Blues_r,antialiased=False)
```

```python
plt.plot(x[0,:],h.real[0,:])
```

# Water mass transformation

```python
nz = 1000
Tz = 1.e1
z_ = np.linspace(-500,nz*Tz,nz/2)

x, y, z = np.meshgrid(x_, y_, z_, indexing='ij')

assert np.all(x[:,0,0] == x_)
assert np.all(y[0,:,0] == y_)
assert np.all(z[0,0,:] == z_)
```

```python
z0 = 50. # m
epsi0 = 1.e-8 # W/kg

def epsi_hab(epsi0,z0,z):
    return epsi0/(1+z/z0)**2

wb = -0.2*epsi_hab(epsi0,z0,z-np.swapaxes(np.tile(h,(nz/2,1,1)),0,2)).real
wb[z<=np.swapaxes(np.tile(h,(nz/2,1,1)),0,2)] = 0

av_wb = np.sum(wb,axis=(0,1))*Tk*Tl

flat_wb = -0.2*epsi_hab(epsi0,z0,z[0,0,:]).real*Tk*Tl*(n/2)**2
flat_wb[z[0,0,:]<0] = 0

plt.plot(av_wb,z[0,0,:])
plt.plot(flat_wb,z[0,0,:])
plt.ylim([-300,600])
```

```python
z0 = 500. # m
epsi0 = 1.e-8 # W/kg

def epsi_hab(epsi0,z0,z):
    return epsi0/(1+z/z0)**2

wb = -0.2*epsi_hab(epsi0,z0,z-np.swapaxes(np.tile(h,(nz/2,1,1)),0,2)).real
wb[z<=np.swapaxes(np.tile(h,(nz/2,1,1)),0,2)] = 0

plt.figure()
plt.pcolor(y[0,:,:]/1000.,z[0,:,:],wb[0,:,:])
plt.colorbar()

plt.figure()
plt.pcolor(x[:,0,:]/1000.,z[:,0,:],wb[:,0,:])
plt.colorbar()

Epsi = -N**(-2)*np.diff(np.sum(wb,axis=(0,1))*Tk*Tl)/Tz

plt.figure()
plt.plot(Epsi,(z[0,0,:-1]+z[0,0,1:])/2.,"r-")

wb_flat = -0.2*epsi_hab(epsi0,z0,z).real
wb_flat[z<=0] = 0
Epsi_flat = -N**(-2)*np.diff(np.sum(wb_flat,axis=(0,1))*Tk*Tl)/Tz
plt.plot(Epsi_flat,(z[0,0,:-1]+z[0,0,1:])/2.,"b-")
plt.plot([0,0],[-500,2000],"k--")
plt.ylim([-250,1000])
```

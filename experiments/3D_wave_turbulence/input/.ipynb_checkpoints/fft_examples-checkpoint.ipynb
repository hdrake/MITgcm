```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.mplot3d import Axes3D
from scipy.fftpack import fftn, ifftn
%matplotlib inline
```

```python
dt = 0.001
period = 0.5
t = np.arange(0,10.,dt)
x = np.sin(2*np.pi/period*t)

x_fft = np.fft.fft(x)
power = np.abs(x_fft)

sample_freq = np.fft.fftfreq(x.size, d=dt)

plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.plot(sample_freq[:t.size//2], power[:t.size//2])
plt.xlabel("frequency")
plt.ylabel("power")
plt.xlim([0,2*(1./period)])

plt.subplot(1,2,2)
plt.plot(t,np.fft.ifft(x_fft))
plt.plot(t,x,"--")
plt.ylabel("x")
plt.xlabel("time")
plt.tight_layout()
```

# FFT2 Example

```python
dx = 2 # grid spacing in x
dy = 1 # grid spacing in y

nx = 400 # number of grid points in x
ny = 500 # number of grid points in y

Lx = dx*nx # width of domain in x
Ly = dy*ny # width of domain in y

ksig = 1./50 # x-wavenumber of topography

k_ = np.linspace(0,1./(2.0*dx),nx//2+1) # spectral grid in x
l_ = np.linspace(0,1./(2.0*dy),ny//2+1) # spectral grid in y
k,l = np.meshgrid(k_,l_) # make spectral grids 2D

zfft = np.zeros_like(k) # initialize array of fourier coefficients
zfft[0,np.argmin(np.abs(ksig-k_))] = 3. # add amplitude at wavenumber of topography

# apply normalization factor
zfft *= ((nx//2+1)*(ny//2+1))

# apply inverse discrete fourier transform
z = np.fft.ifft2(zfft,(ny//2+1,nx//2+1))

x_ = np.linspace(0,nx*dx,nx//2+1)
y_ = np.linspace(0,ny*dy,ny//2+1)
x,y = np.meshgrid(x_,y_)

fig = plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.pcolor(k,l,np.abs(zfft),cmap=plt.get_cmap("Blues_r"))
plt.colorbar(label="power")
plt.xlabel("k")
plt.ylabel("l")

plt.subplot(1,2,2)
plt.pcolor(x,y,z.real,cmap=plt.get_cmap("RdBu_r"))
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
```

```python
dx = 2 # grid spacing in x
dy = 1 # grid spacing in y

nx = 400 # number of grid points in x
ny = 500 # number of grid points in y

Lx = dx*nx # width of domain in x
Ly = dy*ny # width of domain in y

ksig = 1./50 # x-wavenumber of topography

k_ = np.fft.fftfreq(nx//2+1, d=dx) # spectral grid in x
l_ = np.fft.fftfreq(ny//2+1, d=dy) # spectral grid in y
k,l = np.meshgrid(k_,l_) # make spectral grids 2D

zfft = np.zeros_like(k) # initialize array of fourier coefficients
zfft[0,np.argmin(np.abs(ksig-k_))] = 3. # add amplitude at wavenumber of topography

# apply normalization factor
zfft *= ((nx//2+1)*(ny//2+1))

# apply inverse discrete fourier transform
z = np.fft.ifft2(zfft,(ny//2+1,nx//2+1))

x_ = np.linspace(0,nx*dx,nx//2+1)
y_ = np.linspace(0,ny*dy,ny//2+1)
x,y = np.meshgrid(x_,y_)

fig = plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.contourf(k,l,np.abs(zfft),cmap=plt.get_cmap("Blues_r"))
plt.colorbar(label="power")
plt.xlabel("k")
plt.ylabel("l")

plt.subplot(1,2,2)
plt.pcolor(x,y,z.real,cmap=plt.get_cmap("RdBu_r"))
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
```

```python
x_ = np.linspace(0,nx*dx,nx//2+1)
y_ = np.linspace(0,ny*dy,ny//2+1)
x,y = np.meshgrid(x_,y_)
z = 3*np.cos(2*np.pi*ksig*x)

zfft = np.fft.fft2(z,(ny//2+1,nx//2+1))

k_ = np.fft.fftfreq(nx//2+1, d=dx)
l_ = np.fft.fftfreq(ny//2+1, d=dy)
k,l = np.meshgrid(k_,l_)

fig = plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
plt.contourf(k,l,np.abs(zfft),cmap=plt.get_cmap("Blues_r"))
plt.colorbar(label="power")
plt.xlabel("k")
plt.ylabel("l")

# remove normalization factor
zfft /= ((nx//2+1)*(ny//2+1))

plt.subplot(1,2,2)
plt.pcolor(x,y,z.real,cmap=plt.get_cmap("RdBu_r"))
plt.colorbar()
plt.clim([-3,3])
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
```
